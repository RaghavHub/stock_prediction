```python
import json;
from pprint import pprint
```

```python
with open('news_0000001.json') as f:
    data = json.load(f)
```

```python
content = data['text']
```

```python
print(content);
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
ss = sid.polarity_scores(content)
```

```python
print(ss);
```

```python
if ss["compound"] == 0.0: 
        print("Neutral")
elif ss["compound"] > 0.0:
        print("Positive")
else:
        print("Negative")
```
